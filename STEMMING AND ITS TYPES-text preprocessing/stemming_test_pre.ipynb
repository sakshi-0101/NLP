```python
###Stemming
###eating,eat,eaten--->eat
##going,gone,goes--->go
```

```python
words=["eating","eat","eaten","writing","writes","programming","programs","programmer","history"]
```

```python
##PorterStemmer

from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()
```

```python
for word in words:
    print(word+"----->"+stemming.stem(word))
```
