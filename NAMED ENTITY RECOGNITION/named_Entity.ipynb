```python
sentence='Eiffel Tower, wrought-iron structure in Paris that is among the most famous landmarks in the world. It is also a technological masterpiece in building-construction history. It was designed and built (1887–89) by Gustave Eiffel and named in his honor.'
```

```python
import nltk
words=nltk.word_tokenize(sentence)
```

```python
tag_elements=nltk.pos_tag(words)
```

```python
nltk.download("maxent_ne_chunker")
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
```

```python
tree=nltk.ne_chunk(tag_elements)
tree.draw()
```
